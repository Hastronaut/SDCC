1. 数据都是整型或者浮点数，但实际上有0-1变量
2. 49、50列“尿_酸碱度”“尿_葡萄糖”缺失1条，52～56列缺失1600+条，61～76列缺失较多，从几千到一万条数据不等
3. 因变量“癌 ”“癌前病变”“良性疾病”“健康或非结肠疾病”57～60列是没有缺失的0-1变量，这四列可以理解为因变量“得病情况”的亚变量形式。
4. 由上面代码知，数据没有来自重复的病人，故不需考虑纵向数据的情况，可以假设每一行数据都是独立的。

三、数据清洗
3.1 缺失值处理
法1. knn
法2. 数值类型用中位数，分类数据用众数
法3. 用模型预测

具体处理：
1. 49、50列就差一条，用中位数填充
2. 53、54 数值类型，52、61～76 0-1变量。
暂时都采取knn填充，把所有变量都看成数值类型来做

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)  # 假设选择 5 个最近邻
df_complete_knn = imputer.fit_transform(df)
columns=df.columns
index=df.index
df = pd.DataFrame(df_complete_knn, columns=columns, index=index)
```

```python
import seaborn as sns
corrdf = df.corr()
fig = plt.figure(figsize=(20,7))
ax=sns.heatmap(corrdf,vmin=-1,annot=False,cmap="rainbow")
```

```python
#数据比较平衡
print((df["癌"]==1).sum())
print((df["癌前病变"]==1).sum())
print((df["良性疾病"]==1).sum())
print((df["健康或非结肠疾病"]==1).sum())
```

```python
#选择相关性高的变量
v=0.1
a=corrdf['癌'].sort_values(ascending=False)
a=abs(a).sort_values(ascending=False)
c1=a[a>=v]
a=corrdf['癌前病变'].sort_values(ascending=False)
a=abs(a).sort_values(ascending=False)
c2=a[a>=v]
a=corrdf['癌前病变'].sort_values(ascending=False)
a=abs(a).sort_values(ascending=False)
c3=a[a>=v]
a=corrdf['健康或非结肠疾病'].sort_values(ascending=False)
a=abs(a).sort_values(ascending=False)
c4=a[a>=v]

combined_index = c1.index.union(c2.index).union(c3.index).union(c4.index)
excluded_index = ["癌", "癌前病变", "良性疾病", "健康或非结肠疾病"]
selected_columns = combined_index.difference(excluded_index)
selected_columns=selected_columns.tolist()
target_columns=["癌", "癌前病变", "良性疾病", "健康或非结肠疾病"]
```

```python
selected_columns #By correlation
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
X=df[selected_columns]
y = df["癌"]
logistic_regression = LogisticRegression(max_iter=100000)

cv_scores = cross_val_score(logistic_regression, X, y, cv=5, scoring='accuracy')

mean_accuracy = cv_scores.mean()
print("平均准确率:", mean_accuracy)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
X=df.drop(target_columns,axis=1)
y = df["癌"]
logistic_regression = LogisticRegression(max_iter=100000)

# 使用交叉验证计算准确率（假设使用 5 折交叉验证）
cv_scores = cross_val_score(logistic_regression, X, y, cv=5, scoring='accuracy')

mean_accuracy = cv_scores.mean()
print("平均准确率:", mean_accuracy)
```
